```python
#1. PRIME NUMBER

def isprime(n):
    if(n<2):
        print("Not prime")
        return
    for i in range(2,int(n**0.5)+1):
        if (n%i==0):
            print("Not Prime")
            return
    print("Prime")
n=int(input())
isprime(n)
```

```python
#2.PRODUCT OF RANDOM NUMBERS

import random
a=random.randint(1,10)
b=random.randint(1,10)
print(a,"*",b,"=")
n=int(input())
while n!=a*b:
    print("Incorrect, Try again")
    n=int(input())
print("correct")
```

```python
#3.SQUARES OF EVEN OR ODD NUMBERS

print("Squares of even/odd numbers")
a=str(input("Choose even/odd"))
a.lower()
for i in range(100,201):
    if (a=="even" and i%2==0):
            print(f"{i}^2={i*i}")
    elif (a=="odd" and i%2!=0):
        print(f"{i}^2={i*i}")
```

```python
#HOSPITAL BILLING SYSTEM

rc=2000
dc=1500
labtch=300
no_of_days=int(input("number of days stayed by patient="))
no_of_tests=int(input("number of tests taken="))
med_charges=int(input("medical charges="))
bill=((no_of_days*rc)+(no_of_tests*labtch)+med_charges+dc)
discount=bill*0.10
if bill>=10000:
    print(f"Final Bill={bill-discount}")
else:
    print(f"Final Bill={bill}")
```
