```python
#1. Load the patient dataset and show summary

import pandas as pd
import numpy as np

df1=pd.read_csv("Patient_Data.csv")
df2=pd.read_csv("Billing_Data.csv")

print("<=======================Patients data======================>")
print(df1.info)
print("<========================Billing data======================>")
print(df2.info)
```

```python
#2. Select only COLUMNS RELEVANT  to Billing

print("<===========================SPECIFIC COLUMNS=====================>")
df1[["PatientID","Department","Doctor","BillAmount"]]
```

```python
#3. DROP administrative columns

#df1.drop(columns=["ReceptionistID","CheckInTime"],inplace=True)        dataframe itself gets modifies so inplace true
print("<=========================DROP COLUMNS==================================>")
updated=df1.drop(columns=["ReceptionistID","CheckInTime"])                #new variable "updated" holds modified data
updated
```

```python
#4. Total Bill Amount per Department with GroupBy

print("<=========================TOTAL BILL AMOUNT(Dept)==================================>")
df1.groupby("Department")["BillAmount"].sum()
```

```python
#5. Remove duplicate patient records on PatientID

print("<=========================DROP DUPLICATES==================================>")
updated=updated.drop_duplicates("PatientID")
updated
```

```python
#6. Fill Billing Amount missing with its mean

print("<=========================FILLING MISSING VALUES(Mean)==================================>")
updated["BillAmount"]=updated["BillAmount"].fillna(np.round(updated["BillAmount"].mean(),1))
print(updated["BillAmount"])
```

```python
updated
```

```python
#7. Merge the billing dataset with patient dataset on PatientID

print("<=========================MERGE DATASET==================================>")
merged=pd.merge(updated,df2,on="PatientID")
merged
```

```python
#8. Concatinate new patients for the current week(row-wise)

df_new=pd.DataFrame({
    'PatientID':[106,107],
    'Name':['Bunny','Aloo'],
    'Department':['Archeology','History'],
    'Doctor':['Dr. Praneetha','Dr. Pranav'],
    'BillAmount':[4000.0,7500.0],
})
updated_df=pd.concat([merged,df_new],ignore_index=True)
updated_df
```

```python
#9.	Concatenate new category columns like ['InsuranceCovered', 'FinalAmount'] (column-wise).

updated_df.loc[updated_df['PatientID']==106,["InsuranceCovered","FinalAmount"]]=[1000,5000]
updated_df.loc[updated_df['PatientID']==107,["InsuranceCovered","FinalAmount"]]=[1500,4000]
updated_df
```
