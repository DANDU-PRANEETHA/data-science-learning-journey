```python
#A1. OPERATIONS: append, extend, insert, remove, pop, clear, index, count, sort, and reverse


l1=[1,2,3,4,5]
l2=[6]
print(f"List1={l1} and List2={l2}")
l1.append(l2)
print(f"Appending 2 lists :{l1}")
l1.extend(l2)
print(f"Extend :{l1}")
l1.insert(2,10)
print(f"Inserting :{l1}")
l1.remove(1)
print(f"Remove :{l1}")
l1.pop()
l1.pop()
print(f"pop :After popping,list is :{l1}")
print(f"Index :{l1.index(10)}")
print(f"Count :{l1.count(2)}")
l1.sort()
print(f"Sort:{l1}")
l1.reverse()
print(f"Reverse :{l1}")
l1.clear()
print(f"Clear:{l1}")
```

```python
#A7. Number is +ve/-ve ,even/odd, or 0

n=int(input())
if(n>0 and n%2==0):
    print(f"{n} is positive and even")
elif(n>0 and n%2!=0):
    print(f"{n} is positive and odd")
elif(n<0 and n%2==0):
    print(f"{n} is negative and even")
elif(n<0 and n%2!=0):
    print(f"{n} is negative and odd")
else:
    print(f"entered number is 0")
```
